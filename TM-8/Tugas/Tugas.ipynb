```python
import numpy as np
```

```python
def find_fuzzy_value(hw1, hw2, hw3, mid, exam, name):
    # Define the fuzzy sets for homework and exam scores
    homework_low = np.array([0, 30, 60])
    homework_medium = np.array([30, 70, 100])
    homework_high = np.array([70, 100, 100])

    exam_low = np.array([0, 30, 60])
    exam_medium = np.array([30, 70, 100])
    exam_high = np.array([70, 100, 100])

    mid_exam_low = np.array([0, 30, 60])
    mid_exam_medium = np.array([30, 70, 100])
    mid_exam_high = np.array([70, 100, 100])

    # Define the input values (homework and exam scores)
    homework_score = (hw1 + hw2 + hw3) / 3  # For example, homework score of 45
    mid_exam_score = mid     # For example, exam score of 75
    exam_score = exam

    # Calculate the membership values for each input in the fuzzy sets
    homework_low_membership = np.interp(homework_score, homework_low, [0, 1, 0])
    homework_medium_membership = np.interp(homework_score, homework_medium, [0, 1, 0])
    homework_high_membership = np.interp(homework_score, homework_high, [0, 1, 0])

    mid_exam_low_membership = np.interp(mid_exam_score, mid_exam_low, [0, 1, 0])
    mid_exam_medium_membership = np.interp(mid_exam_score, mid_exam_medium, [0, 1, 0])
    mid_exam_high_membership = np.interp(mid_exam_score, mid_exam_high, [0, 1, 0])

    exam_low_membership = np.interp(exam_score, exam_low, [0, 1, 0])
    exam_medium_membership = np.interp(exam_score, exam_medium, [0, 1, 0])
    exam_high_membership = np.interp(exam_score, exam_high, [0, 1, 0])

    # Define the fuzzy rules
    rules = [
        # mid exam low
        (min(homework_low_membership, mid_exam_low_membership, exam_low_membership), "grade_low"),
        (min(homework_medium_membership, mid_exam_low_membership, exam_low_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_low_membership, exam_low_membership), "grade_medium"),
        (min(homework_low_membership, mid_exam_low_membership, exam_medium_membership), "grade_medium"),
        (min(homework_medium_membership, mid_exam_low_membership, exam_medium_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_low_membership, exam_medium_membership), "grade_high"),
        (min(homework_low_membership, mid_exam_low_membership, exam_high_membership), "grade_high"),
        (min(homework_medium_membership, mid_exam_low_membership, exam_high_membership), "grade_high"),
        (min(homework_high_membership, mid_exam_low_membership, exam_high_membership), "grade_high"),
        # mid exam med
        (min(homework_low_membership, mid_exam_medium_membership, exam_low_membership), "grade_low"),
        (min(homework_medium_membership, mid_exam_medium_membership, exam_low_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_medium_membership, exam_low_membership), "grade_medium"),
        (min(homework_low_membership, mid_exam_medium_membership, exam_medium_membership), "grade_medium"),
        (min(homework_medium_membership, mid_exam_medium_membership, exam_medium_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_medium_membership, exam_medium_membership), "grade_high"),
        (min(homework_low_membership, mid_exam_medium_membership, exam_high_membership), "grade_high"),
        (min(homework_medium_membership, mid_exam_medium_membership, exam_high_membership), "grade_high"),
        (min(homework_high_membership, mid_exam_medium_membership, exam_high_membership), "grade_high"),
        # mid exam high
        (min(homework_low_membership, mid_exam_high_membership, exam_low_membership), "grade_low"),
        (min(homework_medium_membership, mid_exam_high_membership, exam_low_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_high_membership, exam_low_membership), "grade_medium"),
        (min(homework_low_membership, mid_exam_high_membership, exam_medium_membership), "grade_medium"),
        (min(homework_medium_membership, mid_exam_high_membership, exam_medium_membership), "grade_medium"),
        (min(homework_high_membership, mid_exam_high_membership, exam_medium_membership), "grade_high"),
        (min(homework_low_membership, mid_exam_high_membership, exam_high_membership), "grade_high"),
        (min(homework_medium_membership, mid_exam_high_membership, exam_high_membership), "grade_high"),
        (min(homework_high_membership, mid_exam_high_membership, exam_high_membership), "grade_high"),
    ]

    # Aggregate the rule outputs
    grades = {
        "grade_low": 0,
        "grade_medium": 0,
        "grade_high": 0,
    }
    
    for rule in rules:
        membership, grade = rule
        grades[grade] = max(grades[grade], membership)

    # Calculate the centroid defuzzified grade
    defuzzified_grade = (grades["grade_low"] * 1 + grades["grade_medium"] * 2 + grades["grade_high"] * 3) / (grades["grade_low"] + grades["grade_medium"] + grades["grade_high"])

    if defuzzified_grade <= 1.5:
        output_grade = "C"
    elif defuzzified_grade < 2.5:
        output_grade = "B"
    else:
        output_grade = "A"
    # Output the recommended grade
    print(name, "Fuzzy Logic Summary: ")
    print("Final Grade :", output_grade)
    print("Grade Value : ", defuzzified_grade)
    print()
```

```python
find_fuzzy_value(80, 85, 75, 80, 40, "Budi")
find_fuzzy_value(70, 80, 75, 80, 40, "Jasmine")
find_fuzzy_value(90, 80, 80, 77, 90, "Rafik")
find_fuzzy_value(50, 55, 60, 10, 30, "Mutia")
find_fuzzy_value(10, 0, 90, 60, 70, "Patrik")
find_fuzzy_value(65, 35, 100, 70, 72, "Fajar")
```
