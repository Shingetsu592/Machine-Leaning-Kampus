```python
import pandas as pd
import scipy as sp
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Open data set 
url = 'data.csv'
# creating data frame
data = pd.read_csv(url)
df = pd.DataFrame(data[['Duration', 'Date', 'Pulse', 'Maxpulse', 'Calories']])
df
```

```python
# Standardize data 
scaler = StandardScaler().fit(df[['Calories']]) 
rescaledX = pd.DataFrame(scaler.transform(df[['Calories']]))
rescaledX.columns = ['Calories Transform']
rescaledX
```
