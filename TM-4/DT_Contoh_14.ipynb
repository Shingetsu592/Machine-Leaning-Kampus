```python
import pandas as pd
import scipy as sp
import numpy as np
from sklearn.preprocessing import MinMaxScaler 
```

```python
# Open data set 
url = 'data.csv'
# creating data frame
data = pd.read_csv(url)
df = pd.DataFrame(data[['Duration', 'Date', 'Pulse', 'Maxpulse', 'Calories']])
df
```

```python
# scaling data
scaler = MinMaxScaler(feature_range=(0 ,1))
rescaledX = pd.DataFrame(scaler.fit_transform(df[['Calories']]))
rescaledX.columns = ['Scaled Calories']
rescaledX
```
