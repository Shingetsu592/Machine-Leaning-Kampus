```python
import pandas as pd
import scipy as sp
import numpy as np
from sklearn.preprocessing import MinMaxScaler, Binarizer, StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Mpg_data.xlsx', sheet_name='auto-mpg', usecols=['mpg', 'cylinders',	'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name'])
df = pd.DataFrame(data)
df
```

```python
# Scaling data
scaler = MinMaxScaler(feature_range=(0 ,1))
rescaledX = pd.DataFrame(scaler.fit_transform(df[['displacement']]))
rescaledX.columns = ['Scaled Displacement']
rescaledX
```

```python
# print scaled displacement plot
plt.plot(rescaledX)
```

```python
# Binarize data
binarizer = Binarizer(threshold = 300.0).fit(df[['mpg']]) 
binaryX = pd.DataFrame(binarizer.transform(df[['mpg']]))
binaryX.columns = ['mpg Binary']
binaryX
```

```python
# print mpg binary with threshold 300
plt.plot(binaryX)
```

```python
# Standardize data 
scaler = StandardScaler().fit(df[['weight']])
scaledX = pd.DataFrame(scaler.transform(df[['weight']]))
scaledX.columns = ['Weight Transform']
scaledX
```

```python
# print transformed weight
plt.plot(scaledX)
```
