```python
import numpy as np
import pandas as pd
import math
import random
import matplotlib.pyplot as plt
```

```python
# random dataset
np.random.seed(5)
n = 5
var1 = np.random.randn(n)
var2 = np.random.randn(n)
var3 = np.random.randn(n)
var4 = np.random.randn(n)
var5 = np.random.randn(n)
```

```python
# output with some math operation
output = var1 - var2 + var3 * var4 / var5 + np.random.randn(n)
print(var1)
print(var2)
print(var3)
print(var4)
print(var5)
print(output)
```

```python
# plot 2 using histogram
plt.plot(var1, color='#793FDF', label= "Variabel 1") # Han Purple
plt.plot(var2, color='#F0B86E', label= "Variabel 2") # Harvest Gold
plt.plot(var3, color='#97FFF4', label= "Variabel 3") # Electric Blue 
plt.plot(var4, color='#CEDEBD', label= "Variabel 4") # Pastel Mint
plt.plot(var5, color='#1450A3', label= "Variabel 5") # sea blue 
plt.plot(output, color='#FF6969', label= "Output") # Bittersweet (pink)
plt.legend()
plt.show()
```

```python
# plot 2 using histogram
plt.hist(var1, color='#793FDF', label= "Variabel 1") # Han Purple
plt.hist(var2, color='#F0B86E', label= "Variabel 2") # Harvest Gold
plt.hist(var3, color='#97FFF4', label= "Variabel 3") # Electric Blue 
plt.hist(var4, color='#CEDEBD', label= "Variabel 4") # Pastel Mint
plt.hist(var5, color='#1450A3', label= "Variabel 5") # sea blue 
plt.hist(output, color='#FF6969', label= "Output") # Bittersweet (pink)  
plt.legend()
plt.show()
```

```python
# plot 3 using event plot
plt.eventplot(var1, orientation="vertical", color='#793FDF',  label= "Variabel 1")
plt.eventplot(var2, orientation="vertical", color='#F0B86E',  label= "Variabel 2")
plt.eventplot(var3, orientation="vertical", color='#97FFF4',  label= "Variabel 3")
plt.eventplot(var4, orientation="vertical", color='#CEDEBD',  label= "Variabel 4")
plt.eventplot(var5, orientation="vertical", color='#1450A3',  label= "Variabel 5")
plt.eventplot(output, orientation="vertical", color='#FF6969', label= "Output")
plt.legend()
plt.show()
```
