```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('data_train_A.csv')
df_train = pd.DataFrame(train)
df_train
# remove timestamp
df_train = df_train.drop(['timestamp'], axis=1)
df_train['Activity'].value_counts()
```

```python
test = pd.read_csv('data_test_A.csv', delimiter=',')
test
# remove timestamp
test = test.drop(['timestamp'], axis=1)
test
```

```python
# select 3000 random data for training
df_train = df_train.sample(n=3000, random_state=42)
# get X_train and y_train
X_train = (df_train.drop(['Activity'], axis=1)).values
y_train = (df_train['Activity']).values
```

```python
class KNNClassifier:
    def __init__(self, k):
        self.k = k
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    
    def euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2)**2))
    
    def predict(self, X_test):
        y_pred = []
        for x in X_test:
            distances = [self.euclidean_distance(x, x_train) for x_train in self.X_train]
            k_indices = np.argsort(distances)[:self.k]
            k_nearest_labels = [self.y_train[i] for i in k_indices]
            most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
            y_pred.append(most_common)
        return y_pred
```

```python
# knn
knn = KNNClassifier(k=2)
knn.fit(X_train, y_train)
y_pred = knn.predict(test.values)
y_pred
```

```python
# create ID and prediction column dataframe
id = np.arange(1, len(y_pred)+1)
id = id.reshape(-1, 1)
id
y_pred = np.array(y_pred)
y_pred = y_pred.reshape(-1, 1)
y_pred
df = np.concatenate((id, y_pred), axis=1)
df = pd.DataFrame(df, columns=['ID', 'label'])
df
```

```python
# save to csv
df.to_csv('submission.csv', index=False)
```
