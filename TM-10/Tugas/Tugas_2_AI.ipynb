```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_excel("AirQualityUCI.xlsx")
df = pd.DataFrame(data)
df
```

```python

# Ambil kolom NMHC dan RH
X = data['NMHC(GT)']
y = data['RH']

# Klasterisasi menggunakan KMeans
kmeans = KMeans(n_clusters=5).fit(X)
labels = kmeans.labels_

# Plot hasil klaster 
plt.scatter(X, y, c=labels, cmap='rainbow')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], color='black', marker='*', label='centroid')

plt.title('Hasil Klasterisasi Data Kualitas Udara')
plt.xlabel('NMHC')
plt.ylabel('RH')
plt.legend()

plt.show()
```
