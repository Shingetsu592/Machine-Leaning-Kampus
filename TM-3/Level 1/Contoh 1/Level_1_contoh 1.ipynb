```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import file from a folder
from os import listdir
FileNames = listdir('pidato')
print(FileNames)
```

```python
pidato_df = pd.DataFrame(index=range(len(FileNames)),columns=['FileName','The Content'])
print(pidato_df)
```

```python
for i,f_name in enumerate(FileNames):
    f = open('pidato/' + f_name, "r", encoding='utf-8')
    f_content = f.readlines()
    f.close()
   
    pidato_df.at[i,'FileName'] = f_name
    pidato_df.at[i,'The Content'] = f_content[0]

pidato_df.columns = ['FileName','The Content']

pidato_df
```
